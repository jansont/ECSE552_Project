```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
training = pd.read_csv('/Users/alixdanglejan-chatillon/ECSE552_Project/Results/TrainingResults1.csv')
validation = pd.read_csv('/Users/alixdanglejan-chatillon/ECSE552_Project/Results/ValidationResults1.csv')
```

```python
def running_mean(x, N):
    cumsum = np.cumsum(np.insert(x, 0, 0)) 
    return (cumsum[N:] - cumsum[:-N]) / float(N)

L = np.array(training['Error'])
L = L[L < 1000]

L = running_mean(L, 100)
L = L[:40]
plt.plot([i for i in range(len(L))], L, label = 'Training')
V = np.array(validation['Error'].iloc[:40])

plt.plot([i for i in range(len(V))], V, label = 'Validation')
plt.legend()
plt.ylabel('MSE Loss'), plt.xlabel('Epochs')
plt.xlabel('Epoch')

#GCN PLOT
```
